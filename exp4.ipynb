```python
# Load Balancing using Round Robin Algorithm

servers = ["Server 1", "Server 2", "Server 3"]

n = int(input("Enter number of client requests: "))

requests = []

for i in range(n):

    req = input("Enter request {}: ".format(i+1))

    requests.append(req)

print("\nLoad Balancing Output:\n")

server_index = 0

for req in requests:

    print(req, "assigned to", servers[server_index])

    server_index = (server_index + 1) % len(servers)
```
