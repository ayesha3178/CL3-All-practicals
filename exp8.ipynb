```python
# Artificial Immune Pattern Recognition
# Structure Damage Classification

# Training Data
# 0 = Undamaged
# 1 = Damaged

training_data = [
    [2, 3, 0],
    [1, 2, 0],
    [8, 9, 1],
    [9, 8, 1]
]

# Input Structure Data
x = int(input("Enter vibration value 1: "))
y = int(input("Enter vibration value 2: "))

# Minimum Distance
min_distance = 999

classification = -1

# Pattern Recognition
for data in training_data:

    distance = ((x - data[0])**2 +
                (y - data[1])**2) ** 0.5

    if distance < min_distance:

        min_distance = distance

        classification = data[2]

# Result
if classification == 0:
    print("\nStructure is UNDAMAGED")
else:
    print("\nStructure is DAMAGED")
```
