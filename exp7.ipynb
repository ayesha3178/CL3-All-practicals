```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

# Load Image
img = Image.open("image.jpg")

# Resize Image
img = img.resize((256, 256))

# Convert to array
img = np.array(img)

# Normalize
img = img / 255.0

# Display Original Image
plt.imshow(img)
plt.title("Original Image")
plt.axis("off")
plt.show()

# Create Artistic Effect
stylized = np.flipud(img)

# Display Stylized Image
plt.imshow(stylized)
plt.title("Stylized Art Image")
plt.axis("off")
plt.show()
```
