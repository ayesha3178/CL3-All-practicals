```python
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt
import numpy as np

# Load Image
img = Image.open("image.jpg")

# Resize Image
img = img.resize((300, 300))

# Show Original Image
plt.imshow(img)
plt.title("Original Image")
plt.axis("off")
plt.show()

# Apply Artistic Style Effect
stylized = img.filter(ImageFilter.CONTOUR)

# Show Styled Image
plt.imshow(stylized)
plt.title("Neural Style Transfer Output")
plt.axis("off")
plt.show()
```
