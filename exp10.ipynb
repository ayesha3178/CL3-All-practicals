```python
import random

# Distance Matrix
distance = [
    [0, 2, 9, 10],
    [1, 0, 6, 4],
    [15, 7, 0, 8],
    [6, 3, 12, 0]
]

cities = [0, 1, 2, 3]

best_path = []
best_distance = 9999

# Simulate Ants
for i in range(10):

    path = cities.copy()

    random.shuffle(path)

    total = 0

    # Calculate Distance
    for j in range(len(path)-1):

        total += distance[path[j]][path[j+1]]

    # Return to starting city
    total += distance[path[-1]][path[0]]

    print("Path:", path, "Distance:", total)

    # Find Best Path
    if total < best_distance:

        best_distance = total

        best_path = path.copy()

print("\nShortest Path:", best_path)
print("Minimum Distance:", best_distance)
```
