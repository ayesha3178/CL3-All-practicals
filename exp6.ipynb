```python
import random

# Initial antibodies (population)
population = [random.randint(1, 100) for i in range(5)]

print("Initial Population:")
print(population)

# Fitness Function
def fitness(x):
    return x

# Select Best Antibody
best = max(population, key=fitness)

print("\nBest Antibody:", best)

# Cloning
clones = [best for i in range(5)]

print("\nClones:")
print(clones)

# Mutation
mutated = []

for c in clones:

    mutation = c + random.randint(-10, 10)

    mutated.append(mutation)

print("\nMutated Clones:")
print(mutated)

# Best after mutation
final_best = max(mutated, key=fitness)

print("\nFinal Best Solution:", final_best)
```
