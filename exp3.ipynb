```python
# Fuzzy Set Operations in Python

n = int(input("Enter number of elements: "))

A = []
B = []

print("\nEnter elements of Fuzzy Set A:")
for i in range(n):
    A.append(float(input()))

print("\nEnter elements of Fuzzy Set B:")
for i in range(n):
    B.append(float(input()))

# UNION
print("\nUnion:")
for i in range(n):
    print(max(A[i], B[i]), end=" ")

# INTERSECTION
print("\n\nIntersection:")
for i in range(n):
    print(min(A[i], B[i]), end=" ")

# COMPLEMENT OF A
print("\n\nComplement of A:")
for i in range(n):
    print(1 - A[i], end=" ")

# DIFFERENCE A-B
print("\n\nDifference A-B:")
for i in range(n):
    print(min(A[i], 1 - B[i]), end=" ")

# CARTESIAN PRODUCT
print("\n\nCartesian Product:")

for i in range(n):
    for j in range(n):
        print(min(A[i], B[j]), end=" ")
    print()

# MAX-MIN COMPOSITION
print("\nMax-Min Composition:")

for i in range(n):
    for j in range(n):

        max_val = 0

        for k in range(n):

            value = min(
                min(A[i], B[k]),
                min(B[k], A[j])
            )

            max_val = max(max_val, value)

        print(max_val, end=" ")

    print()
```
